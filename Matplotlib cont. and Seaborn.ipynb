```python
# First method for subplots
```

```python
x=np.arange(1,10,0.1)
y1=np.sin(x)
y2=np.tan(x)
y3=np.log(x)
y4=np.cos(x)
```

```python
fig,ax=plt.subplots(2,1, figsize=(5,5))
ax[0].plot(x, y1)
ax[1].plot(x, y2)
```

```python
fig,ax=plt.subplots(1,2, figsize=(10,5))
ax[0].plot(x, y1)
ax[1].plot(x, y2)
```

```python
fig,ax=plt.subplots(1,3, figsize=(10,5))
ax[0].plot(x, y1)
ax[1].plot(x, y2)
ax[2].plot(x, y3)
```

```python
fig,ax=plt.subplots(2,2, figsize=(10,10))
ax[0][0].plot(x, y1)
ax[0][1].plot(x, y2)
ax[1][0].plot(x, y3)
ax[1][1].plot(x, y4)
```

```python
fig,ax=plt.subplots(2,2, figsize=(5,5))
ax[0,0].plot(x, y1)
ax[0,1].plot(x, y2)
ax[1,0].plot(x, y3)
ax[1,1].plot(x, y4)
```

```python
fig,ax=plt.subplots(2,2, figsize=(5,5))
ax[0,0].plot(x, y1)
ax[0,0].set_title('sin')
ax[0,1].plot(x, y2)
ax[0,1].set_title('tan')
ax[1,0].plot(x, y3)
ax[1,0].set_title('log')
ax[1,1].plot(x, y4)
ax[1,1].set_title('cos')
```

```python
fig,ax=plt.subplots(1,3, figsize=(8,10))
ax[0].plot(x, y1)
ax[0].set_title('sin')
ax[0].set(xlabel='x val', ylabel='y val')
ax[1].plot(x, y2)
ax[1].set(xlabel='x val', ylabel='y val')
ax[1].set_title('tan')
ax[2].plot(x, y3)
ax[2].set(xlabel='x val', ylabel='y val')
ax[2].set_title('log')
# ax[1,1].plot(x, y4)
# ax[1,1].set(xlabel='x val', ylabel='y val')
# ax[1,1].set_title('cos')
```

```python
fig,ax=plt.subplots(2,2, figsize=(8,10))
ax[0,0].plot(x, y1)
ax[0,0].set_title('sin')
ax[0,0].set(xlabel='x val', ylabel='y val')
ax[0,1].plot(x, y2)
ax[0,1].set(xlabel='x val', ylabel='y val')
ax[0,1].set_title('tan')
ax[1,0].plot(x, y3)
ax[1,0].set(xlabel='x val', ylabel='y val')
ax[1,0].set_title('log')
ax[1,1].plot(x, y4)
ax[1,1].set(xlabel='x val', ylabel='y val')
ax[1,1].set_title('cos')
```

```python
x=np.arange(1,10,0.1)
y1=np.sin(x)
y2=np.tan(x)
y3=np.log(x)
y4=np.cos(x)
```

```python
plt.figure(figsize=(15,8))
plt.subplot(2,3,1)
plt.plot(x,y1)

plt.subplot(2,3,2)
plt.plot(x,y2)

plt.subplot(2,3,3)
plt.plot(x,y3)

plt.subplot(2,3,4)
plt.plot(x,y4)
```

```python
plt.figure(figsize=(15,8))
plt.subplot(2,3,1)
plt.plot(x,y1)

plt.subplot(2,3,2)
plt.plot(x,y2)

plt.subplot(2,3,4)
plt.plot(x,y3)

plt.subplot(2,3,6)
plt.plot(x,y4)
```

```python
plt.figure(figsize=(8,8))
plt.title('diff graphs')
plt.subplot(2,2,1)
plt.plot(x,y1)

plt.subplot(2,2,2)
plt.plot(x,y2)

plt.subplot(2,2,3)
plt.plot(x,y3)

plt.subplot(2,2,4)
plt.plot(x,y4)
```
